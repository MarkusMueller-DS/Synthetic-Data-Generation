```python
import pandas as pd
```

```python
path_to_file = "../results/results_df_log.csv"
df = pd.read_csv(path_to_file)
```

```python
df.info()
```

```python
df.columns
```

```python
df.drop(columns=["ID", "Timestamp", "RUN"], inplace=True)
```

```python
df
```

```python
df.rename(columns={"LR_roc-auc": "LR_roc_auc"}, inplace=True)
```

```python
# aggregate scores
accuracy_columns = [col for col in df.columns if 'accuracy' in col]
f1_score_columns = [col for col in df.columns if 'f1-score' in col]
roc_auc_columns = [col for col in df.columns if 'roc_auc' in col]
```

```python
df[accuracy_columns] = df[accuracy_columns].replace({',': '.'}, regex=True).astype(float)
df[f1_score_columns] = df[f1_score_columns].replace({',': '.'}, regex=True).astype(float)
df[roc_auc_columns] = df[roc_auc_columns].replace({',': '.'}, regex=True).astype(float)
```

```python
accuracy_avg = df[accuracy_columns].mean(axis=1).round(2)
f1_score_avg = df[f1_score_columns].mean(axis=1).round(2)
roc_auc_avg = df[roc_auc_columns].mean(axis=1).round(2)
```

```python
df["accuracy_avg"] = accuracy_avg
df["f1_score_avg"] = f1_score_avg
df["roc_auc_avg"] = roc_auc_avg
```

```python
df[df["Dataset"] == "adult"].to_excel("../results/adult.xlsx", index=False)
df[df["Dataset"] == "yeast"].to_excel("../results/yeast.xlsx", index=False)
df[df["Dataset"] == "cc-fraud-1"].to_excel("../results/cc-fraud-1.xlsx", index=False)
df[df["Dataset"] == "cc-fraud-5"].to_excel("../results/cc-fraud-5.xlsx", index=False)
```
