```python
import pandas as pd
from ucimlrepo import fetch_ucirepo
```

```python
# fetch dataset
adult = fetch_ucirepo(id=2)

# data
X = adult.data.features
Y = adult.data.targets
```

```python
adult.metadata
```

```python
# combine features and targets
df = pd.concat([X, Y], axis=1)
```

```python
df
```

```python
# save df to csv

df.to_csv("../data/raw/adult.csv", index=False)
```
